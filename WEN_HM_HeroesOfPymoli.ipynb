```python
# Dependencies and Setup
#%autosave 0
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
#purchase_data = purchase_data.style.hide_index()
purchase_data.head()
```

## Player Count

* Display the total number of players


```python

# Display the total number of players
uniquePlayers = purchase_data["SN"].nunique()
totalPlayers = pd.DataFrame({"Total Players":[uniquePlayers]})

#print(f"\nThe total of unique players:\n\n {totalPlayers}\n")
totalPlayers
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
#Purchasing Analysis (Total)
#Number of unique items:
uniqueItems = purchase_data["Item ID"].nunique()
# Average price
averagePrice = round(purchase_data["Price"].mean(), 2)

# Total number of Purchases is the number of Purchase IDs
totalPurchaces = purchase_data["Purchase ID"].nunique()
#Total Revenue
totalRevenue = purchase_data["Price"].sum()

#Make new data frame and dictionary
purchaseAnalysis = pd.DataFrame({"Number of unique Items": [uniqueItems], 
                                "Average Price": [averagePrice],
                                "Number of Purchases": [totalPurchaces], 
                                "Total Revenue": [totalRevenue] })

purchaseAnalysis['Total Revenue'] = purchaseAnalysis['Total Revenue'].map("${:,.2f}".format)

#print(f"Purchace Analysis\n\n{purchaseAnalysis}\n")
purchaseAnalysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
#Gender Demographics
#Make data table with males only... but all of the columns
malesOnly = purchase_data.loc[purchase_data['Gender'] == "Male", :]

# Using the Screen Name... make the malesOnly unique 
uniqueMalesonly = malesOnly['SN'].nunique()

#Now repeat making data set for females only...
femalesOnly = purchase_data.loc[purchase_data['Gender'] == "Female", :]

# Make females unique using Screen Name...
uniqueFemalesonly = femalesOnly['SN'].nunique()

#Other / Non-Disclosed playes
otherPlayers = purchase_data.loc[purchase_data['Gender'] == "Other / Non-Disclosed", :]
uniqueOtherPlayers = otherPlayers['SN'].nunique()

#calculate percent for each group using UniquePlayers
malePercentage = (uniqueMalesonly/uniquePlayers)*100
femalePercentage = (uniqueFemalesonly/uniquePlayers)*100
otherPercentage = (uniqueOtherPlayers/uniquePlayers)*100

#Make new data frame with the new gender data as a dictionary
genderDemographics = pd.DataFrame({'Gender' :['Male', 'Female', 'Other / Non-Disclosed'],
                                   'Total Count' :[uniqueMalesonly, uniqueFemalesonly, uniqueOtherPlayers],
                                    'Pecentage of Players' : [malePercentage, femalePercentage, otherPercentage]})

# df.set_index(['year', 'month'])
genderDemographics = genderDemographics.set_index(['Gender'])
#df.index.name = 'foo'
genderDemographics.index.name = ''
#df.sort_index(inplace=True)

genderDemographics["Pecentage of Players"] = genderDemographics["Pecentage of Players"].map("{:.2f}%".format)  

#print(f"Gender Demographics\n\n{genderDemographics}\n")
genderDemographics
```

```python
#Age Demographics:
#bins by age:
agebins = [0,9,14,19,24,29,34,39,100]
ageLabels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']

# Categorize the existing players using the age bins. Hint: use pd.cut()
purchase_data["Age Groups"] = pd.cut(purchase_data["Age"], agebins, labels=ageLabels)

# Calculate the numbers and percentages by age group
# Lets find the number of players in each age group and make that unique; find percentage
playerslessTen = purchase_data.loc[purchase_data["Age Groups"] == "<10", :]
uniqPlayerslessTen = playerslessTen['SN'].nunique()
percentPlayerslessTen = (uniqPlayerslessTen / uniquePlayers)*100

playersTen_14 = purchase_data.loc[purchase_data["Age Groups"] == "10-14", :]
uniqPlayersTen_14 = playersTen_14['SN'].nunique()
percentPlayersTen_14 = (uniqPlayersTen_14/uniquePlayers)*100

playersFifth_19 = purchase_data.loc[purchase_data["Age Groups"] == "15-19", :]
uniqPlayersFifth_19 = playersFifth_19['SN'].nunique()
percentPlayersFifth_19 = (uniqPlayersFifth_19/uniquePlayers)*100

playersTwenty_24 = purchase_data.loc[purchase_data["Age Groups"] == "20-24", :]
uniqPlayersTwenty_24 = playersTwenty_24['SN'].nunique()
percentPlayersTwenty_24 = (uniqPlayersTwenty_24/uniquePlayers)*100

playersTwenty5_29 = purchase_data.loc[purchase_data["Age Groups"] == "25-29", :]
uniqPlayersTwenty5_29 = playersTwenty5_29['SN'].nunique()
percentPlayersTwenty5_29 = (uniqPlayersTwenty5_29/uniquePlayers)*100

playersThirty_34 = purchase_data.loc[purchase_data["Age Groups"] == "30-34", :]
uniqPlayersThirty_34 = playersThirty_34['SN'].nunique()
percentPlayersThirty_34 = (uniqPlayersThirty_34/uniquePlayers)*100

playersThirty5_39 = purchase_data.loc[purchase_data["Age Groups"] == "35-39", :]
uniqPlayersThirty5_39 = playersThirty5_39['SN'].nunique()
percentPlayersThirty5_39 = (uniqPlayersThirty5_39/uniquePlayers)*100

playersFortyplus = purchase_data.loc[purchase_data["Age Groups"] == "40+", :]
uniqPlayersFortyplus = playersFortyplus['SN'].nunique()
percentPlayersFortyplus = (uniqPlayersFortyplus/uniquePlayers)*100

# Display Age Demographics Table
ageDemographics = pd.DataFrame({ 'Age Groups' :['<10','10-14','15-19','20-24','25-29',
                                                '30-34','35-39','40+'],
                                 'Total Count In Age Group' :[uniqPlayerslessTen,uniqPlayersTen_14,uniqPlayersFifth_19,
                                                              uniqPlayersTwenty_24,uniqPlayersTwenty5_29,uniqPlayersThirty_34,
                                                              uniqPlayersThirty5_39,uniqPlayersFortyplus],
                                 'Percentage of Players' :[percentPlayerslessTen, percentPlayersTen_14, percentPlayersFifth_19,
                                                           percentPlayersTwenty_24,percentPlayersTwenty5_29,percentPlayersThirty_34,
                                                           percentPlayersThirty5_39,percentPlayersFortyplus] })

## index off of AgeGroups?... then hide age groups header??
# df.set_index(['year', 'month'])
ageDemographics = ageDemographics.set_index(['Age Groups'])
#df.index.name = 'foo'
ageDemographics.index.name = ''

ageDemographics['Percentage of Players'] = ageDemographics['Percentage of Players'].map("{:.2f}%".format)
#ageDemographics = ageDemographics.style.hide_index()

#print(f"Age Demographics\n\n{ageDemographics}\n")
ageDemographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Top Spenders
topSpenders = purchase_data[['SN','Price']]

totalTopSpend = topSpenders.groupby('SN')['Price'].sum().to_frame()
totalCountTopSpend = topSpenders.groupby('SN')['Price'].count().to_frame()
averageTopSpend = topSpenders.groupby('SN')['Price'].mean().to_frame()

totalCountTopSpend.columns = ["Purchase Count"]

joinTotalCountATS = totalCountTopSpend.join(averageTopSpend, how="left")
joinTotalCountATS.columns = ["Purchase Count","Average Purchase Price "]

joinTCATSTTSpend = joinTotalCountATS.join(totalTopSpend, how="inner")
joinTCATSTTSpend.columns = ["Purchase Count", "Average Purchase Price","Total Purchase Value"] 

topSpendersGroup = joinTCATSTTSpend[["Purchase Count","Average Purchase Price","Total Purchase Value"]]

topSpendersFinal = topSpendersGroup.sort_values('Total Purchase Value', ascending=False)

topSpendersFinal['Average Purchase Price'] = topSpendersFinal['Average Purchase Price'].map("${:.2f}".format)
topSpendersFinal['Total Purchase Value'] = topSpendersFinal['Total Purchase Value'].map("${:.2f}".format)

#print(f"Top Spenders\n\n{topSpendersFinal.head()}\n")
topSpendersFinal.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
## Most Popular Items
collectedItems = purchase_data.groupby(['Item ID','Item Name'])

collectPurchaseCount = collectedItems['Price'].count()
uniqueItemPrice = collectedItems['Price'].sum()/collectedItems['Price'].count()
totalPurchaseSum = collectedItems['Price'].sum()

mostPopItems = pd.DataFrame({'Purchase Count': collectPurchaseCount, 'Item Price': uniqueItemPrice, 
                             'Total Purchase Value': totalPurchaseSum })

mostPopItemsST = mostPopItems.sort_values(['Purchase Count'], ascending=[False])

mostPopItemsST['Item Price'] = mostPopItemsST['Item Price'].map("${:.2f}".format)
mostPopItemsST['Total Purchase Value'] = mostPopItemsST['Total Purchase Value'].map("${:.2f}".format)

#print(f"Most Popular Items\n\n{mostPopItemsFinal.head()}\n")
mostPopItemsST.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
#Most Profitable Items
mostProfItem = mostPopItems

mostProfItem = mostProfItem.sort_values(['Total Purchase Value'], ascending=[False])
mostProfItem['Total Purchase Value'] = mostProfItem['Total Purchase Value'].map("${:.2f}".format)

#print(f"Most Profitable Items\n\n{mostProfItem.head()}\n")
mostProfItem.head()
```
